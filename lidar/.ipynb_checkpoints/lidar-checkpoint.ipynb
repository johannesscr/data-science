```python
import numpy as np
import json
import matplotlib.pyplot as plt
from PIL import Image
from os import path
import os
```

```python
def file_path(name, ext):
    return f"{path.join(os.getcwd(), name)}.{ext}"


def read_json(file_name):
    with open(file_path(file_name, "json")) as in_file:
        json_content = json.loads(in_file.read())
    return json_content


def depth_map_to_float(json_content):
    return np.array(json_content['depth_data']).astype('float32')


def image_to_float(file_name):
    return np.array(Image.open(file_path(file_name, "jpg")))
```

```python
file_name = "my2D"

# with open(file_path("my2D", "json")) as in_file:
#     json_content = json.loads(in_file.read())
#     depth_map = np.array(json_content['depth_data']).astype('float32')
    
json_content = read_json(file_name)
depth_map = depth_map_to_float(json_content)
    
plt.imshow(depth_map)
plt.colorbar()
```

```python
image = np.array(Image.open(f"{file_path}.jpg"))
plt.imshow(image)
```
